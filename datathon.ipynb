```python
# prompt: import python data science libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
stock_df = pd.read_csv('archive/all_stock_data.csv')
stock_df.head()
```

```python
# prompt: import the ticker_id.csv file from my folder and name the variable ticker_df and show the head of the dataframe

ticker_df = pd.read_csv('archive/Ticker_IDs.csv')
ticker_df.head()
```

```python
# prompt: I want to keep only those ticker in stock_df which has its value in the tickers column of ticker_df

# Extract the tickers from ticker_df
tickers_to_keep = ticker_df['Tickers'].tolist()

# Filter stock_df to keep only rows with tickers present in tickers_to_keep
stock_df_filtered = stock_df[stock_df['Ticker'].isin(tickers_to_keep)]

# Now stock_df_filtered contains only the rows with tickers found in ticker_df
```

```python
stock_df_filtered.head()
```

```python
len(stock_df_filtered)
```

```python
# prompt: import SIC_Codes.csv and name it sic_code_df

sic_code_df = pd.read_csv('archive/SIC_Codes.csv')
sic_code_df.head()
```

```python
ticker_df['sic'].info()
```

```python
sic_code_df.info()
```

```python
ticker_df['sic'] = pd.to_numeric(ticker_df['sic'], errors='coerce').astype('Int64')
```

```python
sic_code_to_ticker_df = pd.merge(ticker_df, sic_code_df, left_on='sic', right_on='SIC Code')
```

```python
sic_code_to_ticker_df.head()
```

```python
sic_code_to_ticker_df.columns
```

```python
sic_code_to_ticker_df = sic_code_to_ticker_df.drop(['index', 'name', 'sic', 'address2'], axis=1)
```

```python
# prompt: final_merged_df = pd.merge(stock_df_filtered, sic_code_to_ticker_df, left_on='Ticker', right_on='Tickers')

final_merged_df = pd.merge(stock_df_filtered, sic_code_to_ticker_df, left_on='Ticker', right_on='Tickers')
```

```python
final_merged_df.head()
```

```python
final_df = final_merged_df.drop(['Tickers'], axis=1)
```

```python
final_df.head()
```

```python
final_df.columns
```

```python
len(final_df)
```

```python
final_df.to_csv('all_stock_data_cleaned.csv', index=False)
```
